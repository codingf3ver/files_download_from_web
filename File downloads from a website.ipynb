```python
import re
import requests
from bs4 import BeautifulSoup as bs
import pandas as pd
pd.set_option('display.max_columns', 500)
```

```python
# get the web content url
r = requests.get('https://aliah.ac.in/').text
```

```python
# converting to Beautifulsoup Object
soup = bs(r, 'lxml')
```

```python
# This will find the angle tag with href as an element and search for file that ends with .pdf
pdf_file_list = soup.body.select('a[href$=".pdf"]')
pdf_file_list
```

```python
#Converting the above list to string so that we can clean up the garbage data
str_list = str(pdf_file_list)
```

```python
#substituting the garbage value with empty string
fresh_link = re.sub(
    '(href=|<|>|Downloads|target="_blank"|]|Download|class="btn btn-info btn-xs"|\+|/a, a|")', "", str_list)
print(fresh_link)
```

```python
#splitting the data for cleaning purpose
data = fresh_link.split('/a, a')
print(data)
```

```python
#filtering unwanted value
filter_data = filter(None, data)
```

```python
#loop through the filtered data
link_list = []
for i in filter_data:
    link_list.append(i.strip('[a/'))#removing [a /a from the list
print(link_list)
```

```python
#Creating dataframe 
df=pd.DataFrame(data=link_list,columns=['Download_File_links'])   
df.head(10)
```
